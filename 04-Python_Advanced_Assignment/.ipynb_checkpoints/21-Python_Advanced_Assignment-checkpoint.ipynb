**Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?**

A probability distribution is a function that describes the probability of a random variable taking on a particular value. It is a mathematical model that can be used to predict the likelihood of different outcomes.

The values in a probability distribution are not meant to be random. They are the result of a mathematical function that has been designed to produce a certain distribution of probabilities.

**Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?**

Yes, there is a distinction between true random numbers and pseudo-random numbers. True random numbers are generated by physical processes that are inherently unpredictable. Pseudo-random numbers are generated by mathematical algorithms that are designed to produce sequences of numbers that appear to be random.

Pseudo-random numbers are considered "good enough" for most applications because they are very difficult to distinguish from true random numbers. In fact, for many applications, the difference between true random numbers and pseudo-random numbers is irrelevant.

**Q3. What are the two main factors that influence the behaviour of a "normal" probability distribution?**

The two main factors that influence the behavior of a normal probability distribution are the mean and the standard deviation. The mean is the average value of the distribution, and the standard deviation is a measure of how spread out the distribution is.

**Q4. Provide a real-life example of a normal distribution.**

One real-life example of a normal distribution is the distribution of heights in a population. If you measure the heights of a large number of people, you will find that the heights are normally distributed. The mean of the distribution will be the average height of the population, and the standard deviation will be a measure of how much variation there is in the heights.

**Q5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows?**

In the short term, a probability distribution will be quite variable. The values will fluctuate up and down, and there will be a lot of uncertainty about what the next value will be. However, as the number of trials grows, the distribution will become more stable. The values will start to cluster around the mean, and the uncertainty will decrease.

**Q8. What is the relationship between exponentiation and logarithms?**

Exponentiation and logarithms are inverse functions. This means that the inverse of the exponential function is the logarithm function, and vice versa.

The exponential function raises a number to a power. For example, the exponential function $e^x$ raises the number $e$ to the power of $x$. The logarithm function takes the logarithm of a number. For example, the logarithm function $\log_e x$ takes the logarithm of $x$ to the base $e$.
