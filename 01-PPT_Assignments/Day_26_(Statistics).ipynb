**1. Scenario: A company wants to analyze the sales performance of its products in different regions. They have collected the following data:  
   Region A: [10, 15, 12, 8, 14]  
   Region B: [18, 20, 16, 22, 25]  
   Calculate the mean sales for each region.**

```python
import numpy as np

# Sales data
region_a = np.array([10, 15, 12, 8, 14])
region_b = np.array([18, 20, 16, 22, 25])

# Calculate mean sales
mean_sales_a = np.mean(region_a)
mean_sales_b = np.mean(region_b)

print("Mean sales for Region A:", mean_sales_a)
print("Mean sales for Region B:", mean_sales_b)
```

**8. Scenario: A company wants to analyze the relationship between employee tenure and job satisfaction. The data collected is as follows:  
   Employee Tenure (in years): [2, 3, 5, 4, 6, 2, 4]  
   Job Satisfaction (on a scale of 1 to 10): [7, 8, 6, 9, 5, 7, 6]  
   Perform a linear regression analysis to predict job satisfaction based on employee tenure.**

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Employee tenure (in years)
employee_tenure = np.array([2, 3, 5, 4, 6, 2, 4])

# Job satisfaction (on a scale of 1 to 10)
job_satisfaction = np.array([7, 8, 6, 9, 5, 7, 6])

# Add a constant term for the intercept
X = sm.add_constant(employee_tenure)

# Fit the linear regression model
model = sm.OLS(job_satisfaction, X)
results = model.fit()

# Plot the regression line
plt.scatter(employee_tenure, job_satisfaction)
plt.plot(employee_tenure, results.predict(), color='red', label='Regression Line')
plt.xlabel('Employee Tenure (in years)')
plt.ylabel('Job Satisfaction')
plt.title('Linear Regression Analysis: Employee Tenure vs Job Satisfaction')
plt.legend()
plt.show()
```

**9. Scenario: A study is conducted to compare the effectiveness of two different medications. The recovery times of the patients in each group are as follows:  
   Medication A: [10, 12, 14, 11, 13]  
   Medication B: [15, 17, 16, 14, 18]  
   Perform an analysis of variance (ANOVA) to determine if there is a significant difference in the mean recovery times between the two medications.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Recovery times for Medication A and Medication B
medication_a = np.array([10, 12, 14, 11, 13])
medication_b = np.array([15, 17, 16, 14, 18])

# Perform ANOVA
f_value, p_value = stats.f_oneway(medication_a, medication_b)

print("ANOVA Results")
print("F-value:", f_value)
print("p-value:", p_value)

# Plot boxplots for visualization
data = [medication_a, medication_b]
labels = ['Medication A', 'Medication B']

plt.boxplot(data, labels=labels)
plt.xlabel('Medication')
plt.ylabel('Recovery Time')
plt.title('Comparison of Recovery Times for Medication A and Medication B')
plt.show()
```

**11. Scenario: A quality control department wants to test the weight consistency of a product. The weights of a sample of products are as follows:  
    [10.2, 9.8, 10.0, 10.5, 10.3, 10.1]  
    Perform a hypothesis test to determine if the mean weight differs significantly from 10 grams.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Weights of the product sample
weights = np.array([10.2, 9.8, 10.0, 10.5, 10.3, 10.1])

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(weights, 10)

print("Hypothesis Test Results")
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Plot the histogram
plt.hist(weights, bins=10, edgecolor='black')
plt.axvline(np.mean(weights), color='red', linestyle='--', label='Sample Mean')
plt.axvline(10, color='green', linestyle='--', label='Population Mean')
plt.xlabel('Weights')
plt.ylabel('Frequency')
plt.title('Weight Consistency of the Product')
plt.legend()
plt.show()
```

**12. Scenario: A company wants to analyze the click-through rates of two different website designs. The number of clicks for each design is as follows:   
    Design A: [100, 120, 110, 90, 95]  
    Design B: [80, 85, 90, 95, 100]  
    Perform a chi-square test to determine if there is a significant difference in the click-through rates between the two designs.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Click-through rates for Design A and Design B
clicks_design_a = np.array([100, 120, 110, 90, 95])
clicks_design_b = np.array([80, 85, 90, 95, 100])

# Perform chi-square test
observed = np.array([clicks_design_a, clicks_design_b])
chi2_statistic, p_value = stats.chisquare(observed)

print("Chi-square Test Results")
print("Chi-square statistic:", chi2_statistic)
print("p-value:", p_value)

# Plot the click-through rates
plt.bar(['Design A', 'Design B'], [np.mean(clicks_design_a), np.mean(clicks_design_b)], color=['blue', 'orange'])
plt.xlabel('Website Design')
plt.ylabel('Average Click-through Rate')
plt.title('Comparison of Click-through Rates for Website Designs')
plt.show()
```

**13. Scenario: A survey is conducted to measure customer satisfaction with a product on a scale of 1 to 10. The data collected is as follows:  
    [7, 9, 6, 8, 10, 7, 8, 9, 7, 8]  
    Calculate the 95% confidence interval for the population mean satisfaction score.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Customer satisfaction ratings
satisfaction_ratings = np.array([7, 9, 6, 8, 10, 7, 8, 9, 7, 8])

# Calculate the confidence interval
confidence_interval = stats.t.interval(0.95, len(satisfaction_ratings)-1, loc=np.mean(satisfaction_ratings), scale=stats.sem(satisfaction_ratings))

print("95% Confidence Interval for Population Mean Satisfaction Score:", confidence_interval)

# Plot the histogram
plt.hist(satisfaction_ratings, bins=10, edgecolor='black')
plt.axvline(np.mean(satisfaction_ratings), color='red', linestyle='--', label='Sample Mean')
plt.xlabel('Satisfaction Ratings')
plt.ylabel('Frequency')
plt.title('Customer Satisfaction with the Product')
plt.legend()
plt.show()
```

**14. Scenario: A company wants to analyze the effect of temperature on product performance. The data collected is as follows:  
    Temperature (in degrees Celsius): [20, 22, 23, 19, 21]  
    Performance (on a scale of 1 to 10): [8, 7, 9, 6, 8]  
    Perform a simple linear regression to predict performance based on temperature.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Temperature (in degrees Celsius)
temperature = np.array([20, 22, 23, 19, 21])

# Performance (on a scale of 1 to 10)
performance = np.array([8, 7, 9, 6, 8])

# Perform simple linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(temperature, performance)

# Predict performance based on temperature
predicted_performance = slope * temperature + intercept

print("Simple Linear Regression Results")
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value ** 2)
print("p-value:", p_value)

# Plot the data and regression line
plt.scatter(temperature, performance, color='blue', label='Data')
plt.plot(temperature, predicted_performance, color='red', label='Regression Line')
plt.xlabel('Temperature (°C)')
plt.ylabel('Performance')
plt.title('Effect of Temperature on Product Performance')
plt.legend()
plt.show()
```

**15. Scenario: A study is conducted to compare the preferences of two groups of participants. The preferences are measured on a Likert scale from 1 to 5. The data collected is as follows:  
    Group A: [4, 3, 5, 2, 4]  
    Group B: [3, 2, 4, 3, 3]  
    Perform a Mann-Whitney U test to determine if there is a significant difference in the median preferences between the two groups.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Preferences for Group A and Group B
preferences_group_a = np.array([4, 3, 5, 2, 4])
preferences_group_b = np.array([3, 2, 4, 3, 3])

# Perform Mann-Whitney U test
statistic, p_value = stats.mannwhitneyu(preferences_group_a, preferences_group_b)

print("Mann-Whitney U Test Results")
print("U statistic:", statistic)
print("p-value:", p_value)

# Plot the preferences
plt.boxplot([preferences_group_a, preferences_group_b], labels=['Group A', 'Group B'])
plt.xlabel('Group')
plt.ylabel('Preferences')
plt.title('Comparison of Preferences between Group A and Group B')
plt.show()
```

**17. Scenario: A study is conducted to compare the performance of three different machine learning algorithms. The accuracy scores for each algorithm are as follows:  
    Algorithm A: [0.85, 0.80, 0.82, 0.87, 0.83]  
    Algorithm B: [0.78, 0.82, 0.84, 0.80, 0.79]  
    Algorithm C: [0.90, 0.88, 0.89, 0.86, 0.87]  
    Perform a Kruskal-Wallis test to determine if there is a significant difference in the median accuracy scores between the algorithms.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Accuracy scores for each algorithm
scores_algorithm_a = np.array([0.85, 0.80, 0.82, 0.87, 0.83])
scores_algorithm_b = np.array([0.78, 0.82, 0.84, 0.80, 0.79])
scores_algorithm_c = np.array([0.90, 0.88, 0.89, 0.86, 0.87])

# Perform Kruskal-Wallis test
statistic, p_value = stats.kruskal(scores_algorithm_a, scores_algorithm_b, scores_algorithm_c)

print("Kruskal-Wallis Test Results")
print("Kruskal-Wallis statistic:", statistic)
print("p-value:", p_value)

# Plot the accuracy scores
labels = ['Algorithm A', 'Algorithm B', 'Algorithm C']
scores = [scores_algorithm_a, scores_algorithm_b, scores_algorithm_c]

plt.boxplot(scores, labels=labels)
plt.xlabel('Algorithm')
plt.ylabel('Accuracy Score')
plt.title('Comparison of Accuracy Scores between Algorithms')
plt.show()
```

**18. Scenario: A company wants to analyze the effect of price on sales. The data collected is as follows:  
    Price (in dollars): [10, 15, 12, 8, 14]  
    Sales: [100, 80, 90, 110, 95]  
    Perform a simple linear regression to predict sales based on price.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Price (in dollars)
price = np.array([10, 15, 12, 8, 14])

# Sales
sales = np.array([100, 80, 90, 110, 95])

# Perform simple linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(price, sales)

# Predict sales based on price
predicted_sales = slope * price + intercept

print("Simple Linear Regression Results")
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value ** 2)
print("p-value:", p_value)

# Plot the data and regression line
plt.scatter(price, sales, color='blue', label='Data')
plt.plot(price, predicted_sales, color='red', label='Regression Line')
plt.xlabel('Price ($)')
plt.ylabel('Sales')
plt.title('Effect of Price on Sales')
plt.legend()
plt.show()
```

**19. Scenario: A survey is conducted to measure the satisfaction levels of customers with a new product. The data collected is as follows:  
    [7, 8, 9, 6, 8, 7, 9, 7, 8, 7]  
    Calculate the standard error of the mean satisfaction score.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Satisfaction levels of customers
satisfaction_levels = np.array([7, 8, 9, 6, 8, 7, 9, 7, 8, 7])

# Calculate the standard error of the mean
standard_error = np.std(satisfaction_levels, ddof=1) / np.sqrt(len(satisfaction_levels))

print("Standard Error of the Mean Satisfaction Score:", standard_error)

# Plot the histogram
plt.hist(satisfaction_levels, bins=10, edgecolor='black')
plt.axvline(np.mean(satisfaction_levels), color='red', linestyle='--', label='Sample Mean')
plt.xlabel('Satisfaction Levels')
plt.ylabel('Frequency')
plt.title('Customer Satisfaction with the New Product')
plt.legend()
plt.show()
```

**20. Scenario: A company wants to analyze the relationship between advertising expenditure and sales. The data collected is as follows:  
    Advertising Expenditure (in thousands): [10, 15, 12, 8, 14]  
    Sales (in thousands): [25, 30, 28, 20, 26]  
    Perform a multiple regression analysis to predict sales based on advertising expenditure.**

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")


# Advertising expenditure (in thousands)
expenditure = np.array([10, 15, 12, 8, 14])

# Sales (in thousands)
sales = np.array([25, 30, 28, 20, 26])

# Add constant column to the input data
expenditure_with_constant = sm.add_constant(expenditure)

# Perform multiple regression analysis
model = sm.OLS(sales, expenditure_with_constant)
results = model.fit()

print("Multiple Regression Results")
print(results.summary())

# Predict sales based on advertising expenditure
predicted_sales = results.predict(expenditure_with_constant)

# Plot the data and regression line
plt.scatter(expenditure, sales, color='blue', label='Data')
plt.plot(expenditure, predicted_sales, color='red', label='Regression Line')
plt.xlabel('Advertising Expenditure (in thousands)')
plt.ylabel('Sales (in thousands)')
plt.title('Relationship between Advertising Expenditure and Sales')
plt.legend()
plt.show()
```
