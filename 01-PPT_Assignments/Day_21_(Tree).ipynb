Question-1 

	You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

	Input:

	        10

	       /   \

	     2      7

	   /   \

	 8      4

	Output:

	        8

	      /   \

	    4     10

	  /   \

	2      7

```python
# Question-1: Convert Binary Tree to Binary Search Tree
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def inorderTraversal(root, arr):
    if root:
        inorderTraversal(root.left, arr)
        arr.append(root.val)
        inorderTraversal(root.right, arr)


def buildBST(arr):
    if not arr:
        return None

    mid = len(arr) // 2
    root = TreeNode(arr[mid])
    root.left = buildBST(arr[:mid])
    root.right = buildBST(arr[mid + 1:])

    return root


def binaryTreeToBST(root):
    if not root:
        return None

    arr = []
    inorderTraversal(root, arr)
    arr.sort()
    return buildBST(arr)


# Input for Question-1
root1 = TreeNode(10)
root1.left = TreeNode(2)
root1.right = TreeNode(7)
root1.left.left = TreeNode(8)
root1.left.right = TreeNode(4)

# Convert Binary Tree to Binary Search Tree
bst_root = binaryTreeToBST(root1)
print("Question-1: Binary Tree converted to Binary Search Tree")

# Function to perform inorder traversal of the BST
def inorder(root):
    if root is None:
        return
    inorder(root.left)
    print(root.val, end=" ")
    inorder(root.right)

# Print the BST (inorder traversal)
print("BST:")
inorder(bst_root)
```

Question-2: 

	Given a Binary Search Tree with all unique values and two keys. Find the distance between two nodes in BST. The given keys always exist in BST.

	Example:

	Consider the following BST:


	View original

	Input-1:

	n = 9

	values = [8, 3, 1, 6, 4, 7, 10, 14,13]

	node-1 = 6

	node-2 = 14

	Output-1:

	The distance between the two keys = 4

	Input-2:

	n = 9

	values = [8, 3, 1, 6, 4, 7, 10, 14,13]

	node-1 = 3

	node-2 = 4

	Output-2:

	The distance between the two keys = 2

```python
# Question-2: Distance between Two Nodes in BST
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def findLCA(root, n1, n2):
    if root is None:
        return None

    if root.val > n1 and root.val > n2:
        return findLCA(root.left, n1, n2)

    if root.val < n1 and root.val < n2:
        return findLCA(root.right, n1, n2)

    return root


def findDistance(root, val):
    if root is None:
        return 0

    if root.val == val:
        return 0

    if val < root.val:
        return 1 + findDistance(root.left, val)

    return 1 + findDistance(root.right, val)


def findDistanceBetweenNodes(root, n1, n2):
    lca = findLCA(root, n1, n2)

    d1 = findDistance(lca, n1)
    d2 = findDistance(lca, n2)

    return d1 + d2


# Input for Question-2
root2 = TreeNode(8)
root2.left = TreeNode(3)
root2.right = TreeNode(10)
root2.left.left = TreeNode(1)
root2.left.right = TreeNode(6)
root2.left.right.left = TreeNode(4)
root2.left.right.right = TreeNode(7)
root2.right.right = TreeNode(14)
root2.right.right.left = TreeNode(13)

node1 = 6
node2 = 14

# Find distance between two nodes in BST
distance = findDistanceBetweenNodes(root2, node1, node2)
print("Distance between the two keys =", distance)
```

	Question-3:

	Write a program to convert a binary tree to a doubly linked list.

	Input:

	        10

	       /   \

	     5     20

	           /   \

	        30     35

	Output:

	5 10 30 20 35

```python
# Question-3: Convert Binary Tree to Doubly Linked List
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def convertToDLL(root):
    if root is None:
        return None

    if root.left:
        left = convertToDLL(root.left)
        while left.right:
            left = left.right
        left.right = root
        root.left = left

    if root.right:
        right = convertToDLL(root.right)
        while right.left:
            right = right.left
        right.left = root
        root.right = right

    return root


def printDLL(head):
    if head is None:
        return

    while head:
        print(head.val, end=" ")
        if head.right:
            print("→", end=" ")
        else:
            print()
        head = head.right


# Input for Question-3
root3 = TreeNode(10)
root3.left = TreeNode(5)
root3.right = TreeNode(20)
root3.right.left = TreeNode(30)
root3.right.right = TreeNode(35)

# Convert Binary Tree to Doubly Linked List
dll_head = convertToDLL(root3)
print("Binary Tree converted to Doubly Linked List")
printDLL(dll_head)
```

	Question-4:

	Write a program to connect nodes at the same level.

	Input:

	        1

	      /   \

	    2      3

	  /   \   /   \

	4     5 6    7

	Output:

	1 → -1

	2 → 3

	3 → -1

	4 → 5

	5 → 6

	6 → 7

	7 → -1

```python
# Question-4: Connect Nodes at the Same Level
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        self.next = None


def connectNodes(root):
    if root is None:
        return None

    queue = [root]

    while queue:
        level_size = len(queue)

        for i in range(level_size):
            node = queue.pop(0)

            if i < level_size - 1:
                node.next = queue[0]

            if node.left:
                queue.append(node.left)

            if node.right:
                queue.append(node.right)

    return root


def printLevelOrder(root):
    if root is None:
        return

    while root:
        curr = root
        while curr:
            print(curr.val, end=" ")
            curr = curr.next
        print()
        root = root.left


# Input for Question-4
root4 = TreeNode(1)
root4.left = TreeNode(2)
root4.right = TreeNode(3)
root4.left.left = TreeNode(4)
root4.left.right = TreeNode(5)
root4.right.left = TreeNode(6)
root4.right.right = TreeNode(7)

# Connect Nodes at the Same Level
connected_root = connectNodes(root4)
print("Nodes connected at the same level:-")
printLevelOrder(connected_root)
```
