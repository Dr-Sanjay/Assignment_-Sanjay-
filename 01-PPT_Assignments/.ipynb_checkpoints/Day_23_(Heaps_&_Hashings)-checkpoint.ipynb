Given preorder of a binary tree, calculate itsÂ The preorder is given as a string with two possible characters.**

1. â€˜lâ€™ denotes the leaf
2. â€˜nâ€™ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a node can â€˜nâ€™ or â€˜lâ€™ or mix of both.

**Examples :**

For the input 'nlnll' with a depth of 2:  
Input  : nlnll  
Output : 2  
<pre>
<code>
        n
       / \
      l   n
         / \
        l   l
</code>
</pre>
For the input 'nlnnlll' with a depth of 3:  
Input  : nlnnlll  
Output : 3  

<pre>
<code>
        n
       / \
      l   n
         / \
        n   l
       / \
      l   l
</code>
</pre>


````python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def constructTree(preorder, preIndex):
    if preIndex[0] >= len(preorder):
        return None

    node = Node(preorder[preIndex[0]])
    preIndex[0] += 1

    if node.data == 'n':
        node.left = constructTree(preorder, preIndex)
        node.right = constructTree(preorder, preIndex)

    return node

def calculateDepth(preorder):
    preIndex = [0]
    root = constructTree(preorder, preIndex)
    return findDepth(root)

def findDepth(node):
    if node is None:
        return -1
    left_depth = findDepth(node.left)
    right_depth = findDepth(node.right)
    return max(left_depth, right_depth) + 1

# Example usage:
preorder = "nlnll"
print("Input:", preorder)
print("Output:", calculateDepth(preorder))
````

**ðŸ’¡ Question-2:
Given a Binary tree, the task is to print the left view of the Binary Tree. The left view of a Binary Tree is a set of leftmost nodes for every level.
Examples:
Input:**
```
        4

      /   \\

    5     2

         /   \\

        3     1

       /  \\

      6    7

```
Output: 4 5 3 6
Explanation:
!https://media.geeksforgeeks.org/wp-content/cdn-uploads/left-view.png
Input:
```
                1

              /   \\

            2       3

             \\

               4

                 \\

                    5

                       \\

                         6

```
Output: 1 2 4 5 6

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def constructTree(arr, root, i, n):
    if i < n:
        temp = Node(arr[i])
        root = temp
        root.left = constructTree(arr, root.left, 2 * i + 1, n)
        root.right = constructTree(arr, root.right, 2 * i + 2, n)
    return root

def leftView(root):
    if root is None:
        return []
    queue = [root]
    left_view = []
    while queue:
        size = len(queue)
        for i in range(size):
            node = queue.pop(0)
            if i == 0:
                left_view.append(node.data)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
    return left_view

# Example usage:
arr = [4, 5, 2, None, None, 3, 1, None, None, None, None, 6, 7]
root = constructTree(arr, None, 0, len(arr))
print("Input Tree:")
print("            4")
print("          /   \\")
print("        5       2")
print("             /   \\")
print("            3     1")
print("           /  \\")
print("          6    7")
print("Output:", leftView(root))
````

**ðŸ’¡ Question-3:
Given a Binary Tree, print the Right view of it.
The right view of a Binary Tree is a set of nodes visible when the tree is visited from the Right side.
Examples:
Input:**
```
     1

  /     \\

```
2         3
/   \       /  \
4     5   6    7
```
         \\

           8

```
Output:
Right view of the tree is 1 3 7 8
Input:
```
     1

   /

8

```
/
7
Output:
Right view of the tree is 1 8 7


````python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def constructTree(arr, root, i, n):
    if i < n:
        temp = Node(arr[i])
        root = temp
        root.left = constructTree(arr, root.left, 2 * i + 1, n)
        root.right = constructTree(arr, root.right, 2 * i + 2, n)
    return root

def rightView(root):
    if root is None:
        return []
    queue = [root]
    right_view = []
    while queue:
        size = len(queue)
        for i in range(size):
            node = queue.pop(0)
            if i == size - 1:
                right_view.append(node.data)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
    return right_view

# Example usage:
arr = [1, 2, 3, None, 4, None, None, None, None, None, 5, None, None, None, None]
root = constructTree(arr, None, 0, len(arr))
print("Input Tree:")
print("         1")
print("       /   \\")
print("     2       3")
print("       \\")
print("         4")
print("           \\")
print("             5")
print("Output:", rightView(root))
````

**ðŸ’¡ Question-4:
Given a Binary Tree, The task is to print the bottom view from left to right. A node x is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.**  
Examples:
Input:
```
         20

       /     \\

    8         22

/      \\         \\

```
5         3        25
```
    /    \\

```
10       14
Output: 5, 10, 3, 14, 25.
Input:
```
         20

       /     \\

    8         22

/      \\      /   \\

```
5         3    4     25
```
     /    \\

 10       14

```
Output:  
5 10 4 14 25.  
Explanation:   
If there are multiple bottom-most nodes for a horizontal distance from the root, then print the later one in the level traversal.
3 and 4 are both the bottom-most nodes at a horizontal distance of 0, we need to print 4.


````python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def constructTree(arr, root, i, n):
    if i < n:
        temp = Node(arr[i])
        root = temp
        root.left = constructTree(arr, root.left, 2 * i + 1, n)
        root.right = constructTree(arr, root.right, 2 * i + 2, n)
    return root

def bottomView(root):
    if root is None:
        return []
    queue = [(root, 0)]
    horizontal_distance_map = {}
    while queue:
        node, hd = queue.pop(0)
        horizontal_distance_map[hd] = node.data
        if node.left:
            queue.append((node.left, hd - 1))
        if node.right:
            queue.append((node.right, hd + 1))
    return list(horizontal_distance_map.values())

# Example usage:
arr = [20, 8, 22, 5, 3, None, 25, None, None, 10, 14]
root = constructTree(arr, None, 0, len(arr))
print("Input Tree:")
print("            20")
print("          /     \\")
print("        8         22")
print("     /    \\         \\")
print("    5       3        25")
print("         /   \\")
print("        10   14")
print("Output:", bottomView(root))
````
