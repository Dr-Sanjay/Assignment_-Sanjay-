## Task 1: Calculate Area with Conditions  

Write a Python function calculate_area that takes two parameters: length and width.It
should calculate and return the area of a rectangle.  
However, add a condition: if the length
is equal to the width, return "This is a square!" instead of the area.  
Then, write a program to
input values for length and width from the user and call the calculate_area function to
display either the area or the message.

```python
# Define the calculate_area function
def calculate_area(length, width):
    if length == width:
        return "This is a square!"
    else:
        return length * width

# Input values for length and width from the user
length = float(input("Enter the length: "))
width = float(input("Enter the width: "))

# Call the calculate_area function and display the result
result = calculate_area(length, width)

# Check if it's a square or an area
if isinstance(result, str):
    print(result)
else:
    print(f"The area of the rectangle is {result} square units.")
```

## Task 2: Generate Fibonacci Series  
Problem Statement:
Write a Python program that generates the Fibonacci sequence up to a specified number of
terms, n.  
The Fibonacci sequence starts with 0 and 1, and each subsequent number in the
sequence is the sum of the two preceding numbers (e.g., 0, 1, 1, 2, 3, 5, 8, ...).  
Prompt the
user to enter the number of terms (n) they want in the sequence and then display the
Fibonacci sequence up to that number of terms.

```python
# Function to generate Fibonacci sequence up to n terms
def generate_fibonacci(n):
    # Initialize the first two terms of the sequence
    fibonacci_sequence = [0, 1]

    # Generate the Fibonacci sequence
    while len(fibonacci_sequence) < n:
        next_term = fibonacci_sequence[-1] + fibonacci_sequence[-2]
        fibonacci_sequence.append(next_term)

    return fibonacci_sequence

# enter the number of terms for which fibonacci sequence required
n = int(input("Enter the number of terms for the Fibonacci sequence: "))

# Check if n is non-negative
if n <= 0:
    print("Please enter a positive integer.")
else:
    # Generate and display the Fibonacci sequence
    fibonacci_sequence = generate_fibonacci(n)
    print("Fibonacci Sequence:")
    print(fibonacci_sequence)
```

## Task 3: MySQL Database Operations with Python ( Compulsory )   

Problem Statement:
Your task is to write a Python program that accomplishes the following:  

First create a database , table and add these column ‘student_id’, ‘first_name’, ‘last_name’,
‘age’, ‘grade’.  

Connects to your MySQL database with python.  

Inserts a new student record into the "students" table with the following details:
First Name: "Alice"
Last Name: "Smith"
Age: 18
Grade: 95.5
Updates the grade of the student with the first name "Alice" to 97.0.  

Deletes the student with the last name "Smith."
Fetches and displays all student records from the "students" table.

```python
import mysql.connector
```

```python
# Function to create a database all credentials are taken from main method in last part of code which can be updated as per user
def create_database(host, username, password, database_name):
    try:
        # Connect to MySQL server
        connection = mysql.connector.connect(
            host=host,
            user=username,
            password=password
        )

        # Create a cursor object
        cursor = connection.cursor()

        # SQL query to create a new database
        create_database_query = f"CREATE DATABASE {database_name}"

        # Execute the query
        cursor.execute(create_database_query)
        print(f"Database '{database_name}' created successfully")

    except mysql.connector.Error as err:
        print(f"Error: {err}")
    
    finally:
        # Close cursor and connection
        if cursor:
            cursor.close()
        if connection.is_connected():
            connection.close()

# Function to create a table
def create_table(host, username, password, database_name):
    try:
        # Connect to MySQL server
        connection = mysql.connector.connect(
            host=host,
            user=username,
            password=password,
            database=database_name  # Specify the database to use
        )

        # Create a cursor object
        cursor = connection.cursor()

        # SQL query to create a table with the specified columns
        create_table_query = """
        CREATE TABLE students (
            student_id INT AUTO_INCREMENT PRIMARY KEY,
            first_name VARCHAR(255),
            last_name VARCHAR(255),
            age INT,
            grade FLOAT(2)
        )
        """

        # Execute the query to create the table
        cursor.execute(create_table_query)
        print("Table 'students' created successfully")

    except mysql.connector.Error as err:
        print(f"Error: {err}")
    
    finally:
        # Close cursor and connection
        if cursor:
            cursor.close()
        if connection.is_connected():
            connection.close()

# Function to insert a new student record
def insert_student_record(host, username, password, database_name, first_name, last_name, age, grade):
    try:
        # Connect to MySQL server
        connection = mysql.connector.connect(
            host=host,
            user=username,
            password=password,
            database=database_name  # Specify the database to use
        )

        # Create a cursor object
        cursor = connection.cursor()

        # SQL query to insert a new student record
        insert_query = """
        INSERT INTO students (first_name, last_name, age, grade)
        VALUES (%s, %s, %s, %s)
        """

        # Data to be inserted
        data = (first_name, last_name, age, grade)

        # Execute the query with data
        cursor.execute(insert_query, data)
        connection.commit()
        print("New student record inserted successfully")

    except mysql.connector.Error as err:
        print(f"Error: {err}")
    
    finally:
        # Close cursor and connection
        if cursor:
            cursor.close()
        if connection.is_connected():
            connection.close()

# Function to update a student's grade
def update_student_grade(host, username, password, database_name, first_name, new_grade):
    try:
        # Connect to MySQL server
        connection = mysql.connector.connect(
            host=host,
            user=username,
            password=password,
            database=database_name  # Specify the database to use
        )

        # Create a cursor object
        cursor = connection.cursor()

        # SQL query to update a student's grade
        update_query = """
        UPDATE students
        SET grade = %s
        WHERE first_name = %s
        """

        # Data for the update
        data = (new_grade, first_name)

        # Execute the update query with data
        cursor.execute(update_query, data)
        connection.commit()
        print(f"Grade updated for student {first_name} successfully")

    except mysql.connector.Error as err:
        print(f"Error: {err}")
    
    finally:
        # Close cursor and connection
        if cursor:
            cursor.close()
        if connection.is_connected():
            connection.close()

# Function to delete a student by last name
def delete_student_by_last_name(host, username, password, database_name, last_name):
    try:
        # Connect to MySQL server
        connection = mysql.connector.connect(
            host=host,
            user=username,
            password=password,
            database=database_name  # Specify the database to use
        )

        # Create a cursor object
        cursor = connection.cursor()

        # SQL query to delete a student by last name
        delete_query = """
        DELETE FROM students
        WHERE last_name = %s
        """

        # Data for the delete operation
        data = (last_name,)

        # Execute the delete query with data
        cursor.execute(delete_query, data)
        connection.commit()
        print(f"Student with last name {last_name} deleted successfully")

    except mysql.connector.Error as err:
        print(f"Error: {err}")
    
    finally:
        # Close cursor and connection
        if cursor:
            cursor.close()
        if connection.is_connected():
            connection.close()

# Function to fetch and display all student records
def fetch_all_students(host, username, password, database_name):
    try:
        # Connect to MySQL server
        connection = mysql.connector.connect(
            host=host,
            user=username,
            password=password,
            database=database_name  # Specify the database to use
        )

        # Create a cursor object
        cursor = connection.cursor()

        # SQL query to fetch all student records
        fetch_query = "SELECT * FROM students"

        # Execute the query
        cursor.execute(fetch_query)

        # Fetch and display the records
        for row in cursor.fetchall():
            print(row)

    except mysql.connector.Error as err:
        print(f"Error: {err}")
    
    finally:
        # Close cursor and connection
        if cursor:
            cursor.close()
        if connection.is_connected():
            connection.close()


if __name__ == "__main__":
    host = "localhost"  # Change to your MySQL server's hostname or IP
    username = "root"  # Replace with your MySQL username
    password = "562311"  # Replace with your MySQL password
    database_name = "School"  # Replace with the desired database name

    # Create the database and table
    create_database(host, username, password, database_name)
    create_table(host, username, password, database_name)

    # Insert a new student record
    insert_student_record(host, username, password, database_name, "Alice", "Smith", 18, 95.5)

    # Update Alice's grade
    update_student_grade(host, username, password, database_name, "Alice", 97.0)

    # Delete the student with the last name "Smith"
    delete_student_by_last_name(host, username, password, database_name, "Smith")

    # Fetch and display all student records
    fetch_all_students(host, username, password, database_name)
```
